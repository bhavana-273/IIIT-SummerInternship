```python
# ✅ Step 1: Install Ultralytics
!pip install ultralytics --upgrade -q

# ✅ Step 2: Import and check Ultralytics version
from ultralytics import YOLO
import ultralytics
ultralytics.checks()

# ✅ Step 3: Download the African Wildlife Dataset YAML
# It is already available online, so we reference it directly
dataset_yaml = "https://raw.githubusercontent.com/ultralytics/ultralytics/main/ultralytics/cfg/datasets/african-wildlife.yaml"

# ✅ Step 4: Load and train the YOLOv8 model
model = YOLO('yolov8n.pt')  # You can also try yolov8s.pt or yolov8m.pt

# Train the model for 15 epochs (you can increase to 25 for better accuracy)
results = model.train(data=dataset_yaml, epochs=15, imgsz=640)

# ✅ Step 5: Evaluate and save results
metrics = model.val()

# ✅ Step 6: View Results Folder
import os
os.listdir("runs/detect/train")
```
